## üîπ Mais sele√ß√µes condicionais
**A partir de agora vamos expandir mais nossas an√°lises para obtermos resultados diversos!! Ent√£o ser√° necess√°rio que voc√™ escolha um emprego ou uma √°rea que voc√™ queira analisar mais profundamente**

üîπAlgumas √°reas que d√£o certo (voc√™ pode escolher outra se quiser): **Data, Software, Engineer, Manager, Electrical Engineer.**

üîπAlgumas √°reas que tem poucos valores, mas tamb√©m d√£o certo s√£o: **DevOps, Front End, Security.**


```python
#Escolha uma √°rea e passe como par√¢metro para o m√©todo contains
empregos_selecionados = df.job.str.contains("Security")
```

## üîπ Criando uma nova coluna
üîπ Vamos criar uma nova coluna que diz se um emprego √© ou n√£o √© um dos selecionados anteriormente.

```python
#Crie a nova coluna que diz se um emprego √© da √°rea escolhida ou n√£o.
nome_da_coluna = "security_jobs"
df[nome_da_coluna] = empregos_selecionados
```

# üîç Encontrando nossos resultados
üîπ Para encontrar nosso resultado, precisamos apenas fazer um groupby, agrupando tanto por ano quanto pela nossa nova coluna!


```python
# Definindo as colunas alvo.

colunas_alvo = ['year', 'salary', 'security_jobs']
```

```python
# Criando nosso grupo.

media_salarial = df[colunas_alvo].groupby(['year', 'security_jobs']).mean()
media_salarial
```

```python
media_salarial.reset_index(inplace = True)
```

```python
media_salarial
```

# üîç Criando visualiza√ß√µes
üîπ Chegou a hora de criar nosso primeiro gr√°fico

üîπPara criar um bom gr√°fico precisamos seguir alguns passos como:


*   Definir um tamanho para a figura
*   Criar o gr√°fico.

*   Arrumar os limites
*   Nomear os eixos e o gr√°fico


```python
#Esse c√≥digo √© para grafico linha
sns.set_style("dark")
plt.figure(figsize = (12, 6))
grafico = sns.lineplot(data = media_salarial, x = 'year', y = 'salary', hue = 'security_jobs')
grafico.set_title("Compara√ß√£o de m√©dia salarial e dados", fontsize = 20)
grafico.set_xlabel("Ano", fontsize = 12)
grafico.set_ylabel("Sal√°rio", fontsize = 12)
```

```python
sns.set_style("dark")
plt.figure(figsize = (12, 6))
grafico = sns.barplot(data = media_salarial, x = 'year', y = 'salary', hue = 'security_jobs')
grafico.set_title("Compara√ß√£o de m√©dia salarial e dados", fontsize = 20)
grafico.set_xlabel("Ano", fontsize = 12)
grafico.set_ylabel("Sal√°rio", fontsize = 12)
grafico.set_ylim(0, 150000)
```
